```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
genres =  ['episode','omnibus','story']
tot_title = []
tot_score = []
#tot_genre = []
for genre in genres:
    #url = f'https://comic.naver.com/webtoon/genre?genre={genre}' # 장르별 순위
    url = f'https://comic.naver.com/webtoon/genre?order=StarScore&view=image&genre={genre}'
    html = requests.get(url)
    soup = BeautifulSoup(html.content,'html.parser')
    print(genre)
    for i in soup.find_all('ul',{'class':'img_list'}):
        tot_title.append([i.find_all('li')[n].dt.string for n in range(0,20)]) #제목
        tot_score.append([i.find_all('li')[x].strong.string for x in range(0,20)]) #점수
        
pd.DataFrame({"에피소드":tot_title[0],
             "옴니버스":tot_title[1],
              "스토리":tot_title[2]
             })        
```
