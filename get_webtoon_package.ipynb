```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

옴니버스 별점순으로 상위 20개 (제목,작가,평점) 추출해 CSV 형식으로 저장

```python
url = 'https://comic.naver.com/webtoon/genre?order=StarScore&view=image&genre=omnibus'
html = requests.get(url)
soup = BeautifulSoup(html.content,'html.parser')
web = soup.find('ul',{'class':'img_list'})
score = list(web.find_all('li')[i].strong.get_text() for i in range(20)) # 점수
title = list(web.find_all('li')[i].dt.get_text() for i in range(20)) # 제목
author = list(web.find_all('li')[i].dd.a.get_text() for i in range(20)) # 작가

webtoon_info = pd.DataFrame({'title':title,
                'score':score,
                'author':author})
```

```python
webtoon_info.to_csv('webtoon_to_csv.csv',index=False)
```
